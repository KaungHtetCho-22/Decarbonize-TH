```python
import pandas as pd 
import numpy as np 

train_df = pd.read_csv('train_clean.csv')
test_df = pd.read_csv('test_clean.csv')
val_df = pd.read_csv('val_clean.csv')
```

```python
feature_cols = train_df.columns.drop(['co2','country'])
feature_cols
```

```python
target_col = ['co2']
```

```python
log_transform_cols = [
    'population', 'gdp']

for col in log_transform_cols:
    train_df[col] = np.log1p(train_df[col])
    val_df[col] = np.log1p(val_df[col])
    test_df[col] = np.log1p(test_df[col])
```

```python
X_train = train_df[feature_cols]
y_train = train_df[target_col]

X_val = val_df[feature_cols]
y_val = val_df[target_col]

X_test = test_df[feature_cols]
y_test = test_df[target_col]
```

```python
X_train.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler

feature_scaler = StandardScaler()

X_train_scaled = feature_scaler.fit_transform(X_train)
X_val_scaled = feature_scaler.transform(X_val)
X_test_scaled = feature_scaler.transform(X_test)
```

```python
import joblib

joblib.dump(feature_scaler, 'feature_scaler.pkl')
```

```python
X_train_scaled
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

# === Train XGBoost ===
xgb_model = XGBRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5,
    random_state=42
)

xgb_model.fit(X_train_scaled, y_train.to_numpy().ravel())  

y_pred = xgb_model.predict(X_val_scaled)

# === Evaluation Metrics ===
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)

print(f"XGBoost Validation Performance:")
print(f"• MSE  = {mse:.2f}")
print(f"• RMSE = {rmse:.2f}")
print(f"• MAE  = {mae:.2f}")
print(f"• R²   = {r2:.4f}")
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

# === Train XGBoost ===
xgb_model = XGBRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5,
    random_state=42
)

xgb_model.fit(X_train_scaled, y_train.to_numpy().ravel())  

y_pred = xgb_model.predict(X_test_scaled)

# === Evaluation Metrics ===
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"XGBoost Validation Performance:")
print(f"• MSE  = {mse:.2f}")
print(f"• RMSE = {rmse:.2f}")
print(f"• MAE  = {mae:.2f}")
print(f"• R²   = {r2:.4f}")
```

```python
import pandas as pd
import numpy as np
import joblib
from xgboost import XGBRegressor

# === Save the trained model ===
joblib.dump(xgb_model, "xgboost_model.joblib")

# === Reload the model and scaler ===
model = joblib.load("xgboost_model.joblib")
scaler = joblib.load("feature_scaler.pkl")

# === Load test data and preprocess ===
test_df = pd.read_csv("test_clean.csv")

log_transform_cols = ['population', 'gdp']
for col in log_transform_cols:
    test_df[col] = np.log1p(test_df[col])

# === Define features and filter Thailand 2022 ===
feature_cols = test_df.columns.drop(['co2', 'country'])
th_2022 = test_df[(test_df['year'] == 2022) & (test_df['country'] == 'Thailand')]

X_th = th_2022[feature_cols]
y_th_true = th_2022['co2'].values[0]

# === Scale and predict ===
X_th_scaled = scaler.transform(X_th)
y_th_pred = model.predict(X_th_scaled)[0]

# === Output the result ===
print(f"Thailand 2022 — True CO₂: {y_th_true:.3f}, Predicted CO₂: {y_th_pred:.3f}, Error: {y_th_pred - y_th_true:.3f}")
```
